```python
import numpy as np
import matplotlib.pyplot as plt

# Crear malla
x = np.linspace(-3.5, 3.5, 200)
y = np.linspace(-3.5, 3.5, 200)
X, Y = np.meshgrid(x, y)

# Funcin sombrero mexicano
F = (1 - (X**2 + Y**2)) * np.exp(-(X**2 + Y**2) / 2)

# Grfico con relleno + lneas + etiquetas
plt.figure(figsize=(16, 12))
#Relleno con colores
cf = plt.contourf(X, Y, F, levels=25, cmap='magma')
#Lneas de contorno
cs = plt.contour(X, Y, F, levels=25, colors='black', linewidths=0.6)

# Personalizacin
plt.title("Contour Plot de la funcin Sombrero Mexicano")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(cf, label="F(x, y)")
plt.tight_layout()
plt.show()
```
